## Matrix multiplication
two main of perfoming multiplication in neural networks and deep learning:
1. Element-wise multiplication
2. Matrix mutliplication (dot product)

there are two main rule performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match:
* `(3, 2) @ (3, 2)` won't work
* `(3, 2) @ (2, 3)` will work
* `(2, 3) @ (3, 2)` will work
2. The resulting matrix has the shape of the **outer 
* `(3, 2) @ (2, 3)` --> `(3, 3)`
* `(2, 3) @ (3, 2)` --> `(2, 2)`

```python
#Import Pytorch and matplotlib
import torch
from torch import nn
import matplotlib.pyplot as plt

#Check PyTorch version
torch.__version__
```

```python
#Setup device agnostic code
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f'Using device : {device}')
```

```python
#Create some data usingi the linear regression formula of y = weights * x + bias 
weight = 0.7
bias = 0.3

#create ranage values
start = 0
end = 1
step = 0.02

#Create X and y (feateures and labels)
X = torch.arange(start, end, step).unsqueeze(dim=1) #without unsqueeze it will be 1D tensor
y = weight * X + bias
X[:10], y[:10]
```

# Split Train & Test

```python
#Split data into training and test sets
train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
#Plot the data
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=None):
    plt.figure(figsize=(10,7))
    #Plot training data in blue
    plt.scatter(train_data, train_labels, c='b', s=4, label='Training data')
    #Plot test data in green
    plt.scatter(test_data, test_labels, c='g', s=4, label='Testing data')
    #Are there predictions?
    if predictions is not None:
        #Plot the predictions in red
        plt.scatter(test_data, predictions, c='r', s=4, label='Predictions')
    #Show the legend
    plt.legend(prop={'size':14})

plot_predictions()
```

# Building a Pytorch Linear Model

```python
#Building a linear regression model

# By inheriting nn.Module (super().__init__()), you get:
# - model.parameters()  all trainable parameters
# - model.to(device)  move to CPU / GPU
# - model.train() / model.eval()
# - automatic gradient tracking
# - ability to use optimizers

#nn.Parameter (nn.Parameter  makes w and b trainable parameter)
# - include it in model.parameters()
# - update it during optimizer step

class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_layer = nn.Linear(in_features=1, out_features=1)  # y = mx + b (m=weight, b=bias) / feature_in is X and feature_out is y
        # self.weights = nn.Parameter(torch.randn(1, #Creates a tensor with random value (random at start)
        #                                     requires_grad=True, #compute gradients, update it during optimization, optimized using gradient descent
        #                                     dtype=torch.float)) #float tensor

        # self.bias = nn.Parameter(torch.randn(1,
        #                                  requires_grad=True,
        #                                  dtype=torch.float))
    
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(x)
        # return self.weights * x + self.bias
```

```python
#set the manual seed
torch.manual_seed(42)
model = LinearRegressionModel()
model, model.state_dict()  # state_dict - model's parameters (weights and bias)
```

```python
#Check the model current device
next(model.parameters()).device
```

```python
#Set the model to the target device
model.to(device)
next(model.parameters()).device
```

# Training
- Loss function
- Optimizer
- Training loop
- Testing loop

```python
#Setuo loss function
loss_fn = nn.L1Loss()  # MAE

#Setup optimizer
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)  # Stochastic Gradient Descent
# model.parameters()  the weights and biases to update
# lr=0.01  learning rate (how big each update step is) (hyperparameter)
```

```python
#Training loop
torch.manual_seed(42)
epochs = 200
#number of epochs - number of steps through the entire dataset

for epoch in range(epochs):
    model.train()  # put the model in training mode

    #1. Forward pass
    y_pred = model(X_train.to(device))

    #2. Calculate the loss
    loss = loss_fn(y_pred, y_train.to(device))

    #3. Zero gradients
    optimizer.zero_grad()
    #Clears (resets) all previously computed gradients.

    #4. Backward pass
    loss.backward()
    #calculate gradients of each parementer (weight and bias)
    #loss / weights.  derevative of loss with respect to weights
    #loss / bias.     derevative of loss with respect to bias

    #5. Update the weights
    optimizer.step()
    #Updates the models parameters using the computed gradients.
    #new_weight = old_weight - learning_rate  gradient

    model.eval() # puts your model into testing mode. / you are telling PyTorch: I am testing this model now.
    with torch.inference_mode():
        #Make predictions
        test_pred = model(X_test.to(device))
        #Calculate the loss
        test_loss = loss_fn(test_pred, y_test.to(device))

    #Print out the progress every 10 epochs
    if epoch % 10 == 0:
        print(f'Epoch: {epoch} | Loss: {loss.item():.4f} | Test Loss: {test_loss.item():.4f}')
```
